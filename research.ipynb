```python
# keep only selected columns.
df_copy = df[['title', 'overview', 'genres','tagline','vote_average','popularity' , 'release_date' , 'id']]
df_copy.head()
```

```python
def remove_missing(df):
    df = df.drop_duplicates(subset=['title'])
    columns = ['title','vote_average','popularity' , 'release_date']
    df = df.dropna(subset = columns)
    df['overview'] = df['overview'].fillna('')
    df['tagline'] = df['tagline'].fillna('')
    return df

df_copy = remove_missing(df_copy)
df_copy.head(3)
```

```python
import re
# [{'id': 16, 'name': 'Animation'}, converts this into :['Animation' , 'Comedy' , 'Fantasy']
def convert_to_list(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# removing stopwords + lemmatization

def preprocess_text(text):
    text = str(text).lower() # lowercase.
    text = re.sub(r'[^a-zA-Z\s]', '', text) # remove whitespaces.
    words = text.split() # split the text into array
    words = [word for word in words if word not in stop_words] # remove stop_words.
    words = [lemmatizer.lemmatize(word) for word in words] # lemmatization
    return " ".join(words) # join the sentence to make it string


df_copy['overview'] = df_copy['overview'].fillna('').apply(lambda x: x.split())
df_copy['tagline'] = df_copy['tagline'].fillna('').apply(lambda x: x.split())
df_copy['genres'] = df_copy['genres'].apply(convert_to_list)
df_copy['tags'] = df_copy['overview'] + df_copy['genres'] + df_copy['tagline']
df_copy['tags'] = df_copy['tags'].apply(lambda x: " ".join(x).lower() if isinstance(x,list)else "")
df_copy['tags'] = df_copy['tags'].apply(preprocess_text)

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from sklearn.metrics.pairwise import cosine_similarity
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df_copy['tags'])
```

```python
import difflib
df_copy = df_copy.reset_index(drop=True)
indices = pd.DataFrame(df_copy.index , index = df_copy['title']).drop_duplicates()

def find_closest_title(user_input, all_titles, cutoff=0.6):
    matches = difflib.get_close_matches(user_input, all_titles, n=1, cutoff=cutoff)
    return matches[0] if matches else None

def recommend(df_copy, title, n=10):
    # use exact title if available, otherwise try to find a close match
    if title not in indices.index:
        closest = find_closest_title(title, indices.index.tolist())
        if closest is None:
            raise KeyError(f"Title '{title}' not found and no close match available.")
        title_used = closest
    else:
        title_used = title

    idx = int(indices.at[title_used, 0])
    sim_score = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
    similar_idx = sim_score.argsort()[::-1][1:n+1]
    return pd.DataFrame(df_copy['title'].iloc[similar_idx])

recommend(df_copy, 'Avengers', 10)
```

```python
import pickle

pickle.dump(tfidf_matrix,open('tfidf_matrix.pkl','wb'))

pickle.dump(indices,open('indices.pkl','wb'))

df_copy.to_pickle('df.pkl')

pickle.dump(tfidf,open('tfidf.pkl','wb'))
```
